# Visualising data using matplotlib, plotly and vega-lite

## Matplotlib

```python
import io
import pandas as pd
import matplotlib.pyplot as plt

raw = """[
  {"region": "us-west", "events": 120, "latency_ms": 42},
  {"region": "us-east", "events": 95, "latency_ms": 55},
  {"region": "eu", "events": 210, "latency_ms": 38}
]"""

df_json = pd.read_json(io.StringIO(raw))
df_json
```

```python
fig, ax = plt.subplots(figsize=(6, 3))
ax.bar(df_json["region"], df_json["events"], color="#2196f3")
ax.set_title("Events by region (from JSON)")
plt.tight_layout()
plt.show()
```

```python
import numpy as np

df = pd.DataFrame({
    "day": pd.date_range("2026-01-01", periods=14, freq="D"),
    "value": np.random.default_rng(7).integers(10, 100, size=14),
})

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(df["day"], df["value"], marker="o")
ax.set_title("Synthetic series from DataFrame")
plt.xticks(rotation=35, ha="right")
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Generate random data
np.random.seed(42)
df = pd.DataFrame({
    "x": np.random.randn(100),
    "y": np.random.randn(100),
    "category": np.random.choice(["A", "B", "C"], size=100)
})

# Create a scatter plot
fig = px.scatter(
    df, 
    x="x", 
    y="y", 
    color="category",
    title="Plotly Scatter Plot Example"
)

fig
```

## Vega-lite

```python
import altair as alt
import pandas as pd

# Example data
data = pd.DataFrame({
    'category': ['A', 'B', 'C', 'D'],
    'value': [4, 6, 10, 3]
})

# Vega-Lite chart using Altair
chart = alt.Chart(data).mark_bar().encode(
    x='category',
    y='value'
).properties(
    width=400,
    height=300,
    title='Simple Vega-Lite Bar Chart'
)

chart
```

## Visualising dataframes with `itables`

```python
await micropip.install("itables")
import itables
import itables.options as opt

# connected=False makes each show() output self-contained so it works
# in isolated cell output frames (no cross-cell init_notebook_mode needed).
opt.connected = False
```

```python
import pandas as pd
import numpy as np
from itables import show

# Sample DataFrame
np.random.seed(0)
df = pd.DataFrame({
    "Category": np.random.choice(['A', 'B', 'C'], size=20),
    "Value": np.random.randint(1, 100, size=20),
    "Score": np.random.rand(20),
    "Date": pd.date_range("2024-01-01", periods=20)
})

show(df)
```
