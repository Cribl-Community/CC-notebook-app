### Setup\n\nRun the next cell **once** before the rest. It installs **Altair** and **Plotly** only when Plotly is not already bundled (Pyodide may ship Plotly **6.x**); otherwise it installs **Altair** only. `micropip` pulls matching transitive wheels in one resolve. Hosts are allow-listed in the app `proxies.yml`. Pandas and Matplotlib ship with the kernel.

```python
import importlib.util\nimport micropip\n\n# Pyodide may bundle Plotly 6.x; only pin 5.24.1 from PyPI when Plotly is absent.\n# (Plotly 6+ pulls narwhals; 5.24.x is the safe pin for a cold micropip install.)\n_pkgs = ['altair==5.5.0']\nif importlib.util.find_spec('plotly') is None:\n    _pkgs.append('plotly==5.24.1')\nawait micropip.install(_pkgs)
```

```python
import pandas as pd\n\ndf = pd.DataFrame([\n    {'region': 'us-west', 'events': 120, 'latency_ms': 42},\n    {'region': 'us-east', 'events': 95, 'latency_ms': 55},\n    {'region': 'eu', 'events': 210, 'latency_ms': 38},\n    {'region': 'apac', 'events': 140, 'latency_ms': 47},\n])\ndf
```

```python
import matplotlib.pyplot as plt\n\nfig, ax = plt.subplots(figsize=(6, 3))\nax.bar(df['region'], df['events'], color='#2196f3')\nax.set_title('Matplotlib: events by region')\nplt.tight_layout()\nplt.show()
```

```python
import plotly.graph_objects as go\n\nfig = go.Figure()\nfor region, g in df.groupby('region'):\n    fig.add_trace(\n        go.Scatter(x=g['events'], y=g['latency_ms'], mode='markers', name=str(region))\n    )\nfig.update_layout(title='Plotly: events vs latency')\nfig.show()
```

```python
import altair as alt\n\nchart = alt.Chart(df).mark_bar().encode(x='region', y='latency_ms', color='region').properties(\n    width=420,\n    height=280,\n    title='Vega-Lite: latency by region',\n)\nchart.show()
```
