# Matplotlib from JSON and pandas

**Matplotlib** is preloaded with Pyodide. Use pyplot and `plt.show()` to render figures in the output area.


```python
import io
import pandas as pd
import matplotlib.pyplot as plt

raw = """[
  {"region": "us-west", "events": 120, "latency_ms": 42},
  {"region": "us-east", "events": 95, "latency_ms": 55},
  {"region": "eu", "events": 210, "latency_ms": 38}
]"""

df_json = pd.read_json(io.StringIO(raw))
df_json
```

```python
fig, ax = plt.subplots(figsize=(6, 3))
ax.bar(df_json["region"], df_json["events"], color="#2196f3")
ax.set_title("Events by region (from JSON)")
plt.tight_layout()
plt.show()
```

```python
import numpy as np

df = pd.DataFrame({
    "day": pd.date_range("2026-01-01", periods=14, freq="D"),
    "value": np.random.default_rng(7).integers(10, 100, size=14),
})

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(df["day"], df["value"], marker="o")
ax.set_title("Synthetic series from DataFrame")
plt.xticks(rotation=35, ha="right")
plt.tight_layout()
plt.show()
```
