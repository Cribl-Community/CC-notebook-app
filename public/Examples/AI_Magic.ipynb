## AI prompt templating patterns

Try these prompt variants with the AI button:

1. `Summarize what can be visualized from {{ df | describe }} and generate one matplotlib chart.`
2. `Using {{ df | describe }}, generate a chart for response-code distribution and explain anomalies.`
3. `Use {{ df | describe }} to choose the best chart type automatically and justify the choice in comments.`

Then iterate by changing the objective while keeping the same Jinja object context.

## Raw object vs `| describe`

Use `{{ df | describe }}` when:
- the dataframe is large
- you want schema/type/shape context without dumping full data
- you want more stable prompt behavior

Use raw `{{ df }}` when:
- the dataframe is small and you want concrete sample rows in prompt context
- exact values matter more than compact summary

```python
# ### Prompt:
# Generate Python code that visualizes the current dataframe.
# Use raw dataframe context (not summary): {{ df }}
#
# Requirements:
# - inspect available columns and choose one meaningful comparison plot
# - include chart title and axis labels
# - add brief comments explaining why this chart fits the visible data sample
```
