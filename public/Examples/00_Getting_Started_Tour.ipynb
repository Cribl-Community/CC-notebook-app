## Cell toolbar: On/Off and If

The examples below use **`metadata.notebook_app`** in the saved file:

- **Off** — cell is greyed out and skipped by **Run** / **Run All** until you turn **On**.
- **If** — a Python expression; the body runs only when it evaluates to true. After a run, **T** / **F** / **E** shows the last condition outcome.
- A condition error skips the body (same as false) and shows **E**.

```python
%%cribl_search var=tour_df lang=kql dataset=cribl_search_sample earliest=-1h latest=now limit=300
dataset=cribl_search_sample
| sort by _time desc
| limit 300
```

```python
print('rows:', len(tour_df))
print('columns:', list(tour_df.columns)[:10])
tour_df.head(5)
```

```python
import matplotlib.pyplot as plt

if len(tour_df) > 0:
    numeric = tour_df.select_dtypes(include=['number']).columns
    if len(numeric) > 0:
        metric = numeric[0]
        tour_df[metric].head(100).plot(kind='line', title=f'Tour chart: {metric}')
        plt.tight_layout()
        plt.show()
    else:
        print('No numeric columns in this result set yet.')
else:
    print('No rows returned.')
```
