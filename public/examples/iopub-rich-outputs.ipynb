# IOPub rich outputs

This notebook exercises the new event-driven output area:

- `display(...)` for rich `display_data` mime bundles
- `update_display_data` for in-place updates via `display_id`
- `clear_output(wait=True)` for flicker-free updates
- `execute_result` from trailing expressions
- `error` traceback

It mirrors the JupyterLab `OutputArea` semantics.

```python
from IPython.display import display, HTML, Markdown, JSON

display(HTML('<b>HTML output</b> rendered through the mime bundle path.'))
display(Markdown('## Markdown output\n- list item one\n- list item two'))
display(JSON({'name': 'cribl', 'enabled': True, 'rows': [1, 2, 3]}))
```

```python
import time
from IPython.display import display, HTML, clear_output

h = display(HTML('<i>working… 0%</i>'), display_id='loader')
for pct in (10, 25, 50, 75, 100):
    time.sleep(0.1)
    h.update(HTML(f'<b>working… {pct}%</b>'))

clear_output(wait=True)
display(HTML('<span style="color: green">done.</span>'))
```
