# itables demo

[itables](https://github.com/mwouts/itables) renders pandas DataFrames as interactive HTML tables (DataTables.net under the hood). The HTML it emits goes through our `text/html` mime renderer.

```python
import pandas as pd
import numpy as np
from itables import init_notebook_mode, show

init_notebook_mode(all_interactive=False)

rng = np.random.default_rng(7)
N = 50
df = pd.DataFrame({
    'id': range(1, N + 1),
    'name': rng.choice(['alpha', 'beta', 'gamma', 'delta'], size=N),
    'value': rng.normal(loc=10, scale=3, size=N).round(2),
    'active': rng.choice([True, False], size=N, p=[0.7, 0.3]),
})
show(df, paging=True, searching=True)
```
