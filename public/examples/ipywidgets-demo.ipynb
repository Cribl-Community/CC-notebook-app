# ipywidgets demo

ipywidgets emit `application/vnd.jupyter.widget-view+json` mime bundles. This notebook does **not** ship a Comm/runtime to bind those widget views to the kernel — they render as a fallback notice and a `text/plain` representation.

Sliders, buttons, and progress bars still produce a static representation, so this is a great smoke test of the IOPub `display_data` path.

```python
import ipywidgets as w
from IPython.display import display

slider = w.IntSlider(value=42, min=0, max=100, step=1, description='Volume')
button = w.Button(description='Click me')
text = w.Text(value='hello', description='Greeting')

display(w.VBox([slider, button, text]))
```

```python
import ipywidgets as w
import time
from IPython.display import display

p = w.IntProgress(value=0, min=0, max=100, description='Progress:')
h = display(p, display_id='prog')
for v in range(0, 101, 25):
    time.sleep(0.1)
    p.value = v
    h.update(p)

print('progress complete; final value =', p.value)
```
