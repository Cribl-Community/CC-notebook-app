# All example notebooks — staging E2E

Playwright can run **Run All** on every bundled example listed in `public/Examples/manifest.json`:

- `npm run e2e:examples` — all examples except the `@heavy` Anomaly PyOD notebook (single worker).
- `npm run e2e:examples:all` — includes Anomaly PyOD (first run can take tens of minutes).

Use the same `e2e/.env` and `npm run e2e:auth` session as `npm run e2e`. This suite is **opt-in** and is not part of the default `npm run e2e` quick/slow split.

Implementation: `e2e/specs/all-example-notebooks.spec.ts`. Details: `docs/E2E_STAGING.md`.
